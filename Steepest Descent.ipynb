```python
import numpy as np
import math

np.set_printoptions(precision=4)
np.set_printoptions(suppress=True)
```

```python
def grad(x):
    return np.dot(Q, x) - b

def grad_equals_zero(g):
    return not np.any(np.around(g, decimals=4))
```

```python
def steepest_descent(f, x0):
    
    (Q, b) = f
    
    x = [x0]
    print("x[{0}]:\n".format(0), x[0])
    
    g = [grad(x[0])]
    print("g[{0}]:\n".format(0), g[0])
    
    alpha = []
    
    print("\n")
    
    i = 0
    while not grad_equals_zero(g[i]):
        
        alpha.append(np.dot(g[i].T, g[i]).item()/np.dot(np.dot(g[i].T, Q), g[i]).item())
        print("alpha[{0}]: ".format(i), alpha[i])
        
        x.append(x[i] - alpha[i]*g[i])
        print("x[{0}]:\n".format(i+1), x[i+1])
        
        i += 1
        
        g.append(grad(x[i]))
        print("g[{0}]:\n".format(i), g[i])
        
        print("\n")
    
    return x[-1]
```

```python
#Ex. 10.10
Q = np.array([[5, 2], [2, 1]])
b = np.array([[3, 1]]).T
f = (Q, b)

x0 = np.array([[0, 0]]).T

mini = steepest_descent(f, x0)
print("Minimizer =")
print(mini)
```

```python
#Eg. 10.3
Q = np.array([[3, 0, 1], [0, 4, 2], [1, 2, 3]])
b = np.array([[3, 0, 1]]).T
f = (Q, b)
x0 = np.array([[0, 0, 0]]).T

print("\nMinimizer, x* = \n", steepest_descent(f, x0))
```
