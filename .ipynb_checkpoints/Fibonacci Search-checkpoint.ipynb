```python
import math
```

```python
def fibo(n):
    if n==0 or n==1:
        return 1
    
    return fibo(n-1) + fibo(n-2)
```

```python
def no_of_iterations(initial_pt, end_pt, tolerance, epsilon):
    
    n = 0
    while fibo(n+1) < (1+2*epsilon)/(tolerance/(end_pt-initial_pt)):
        n+=1
    
    return n
```

```python
def Fibonacci_search(f, a0, b0, tolerance, epsilon):
    
    a = [a0]
    b = [b0]
    print("a(0) = ", a[0])
    print("b(0) = ", b[0])
    print("\n")
    
    N = no_of_iterations(a[0], b[0], tolerance, epsilon)
    
    rho = [1 - (fibo(N)/fibo(N+1))]
    
    a.append(a[0] + rho[0]*(b[0]-a[0]))
    b.append(a[0] + (1-rho[0])*(b[0]-a[0]))
        
    l = a[0]
    r = b[0]
    i = 1
    while i <= N:
        
        print("a({}) = {}".format(i, a[i]))
        print("b({}) = {}".format(i, b[i]))
        
        f_a = f(a[i])
        f_b = f(b[i])
        
        print("f(a({})) = {}".format(i, f_a))
        print("f(b({})) = {}".format(i, f_b))
        
        if i<N-1:
            rho.append(1 - fibo(N-i)/fibo(N-i+1))
        else:
            rho.append((1-2*epsilon)/2)
        
        print("Uncertainty interval: ")        
        if f_a < f_b:
            r = b[i]
            print("[a({}), b({})] = [{}, {}]".format(a.index(l), b.index(r), l, r))
            b.append(a[i])
            a.append(l + rho[i]*(r-l))
        
        elif f_a >= f_b:
            l = a[i]
            print("[a({}), b({})] = [{}, {}]".format(a.index(l), b.index(r), l, r))
            a.append(b[i])
            b.append(l + (1-rho[i])*(r-l))
        
        print("\n")
        i+=1
```

```python
#Eg. 7.2
def f(x):
    return pow(x,4) - 14*pow(x,3) + 60*pow(x,2) - 70*x

Fibonacci_search(f, a0=0, b0=2, tolerance=0.3, epsilon=0.05)
```

```python
#Ex. 7.2
def f(x):
    return x**2 + 4*math.cos(x)

Fibonacci_search(f, a0=1, b0=2, tolerance=0.2, epsilon=0.05)
```

```python
#Ex 7.3
def f(x):
    return 8*math.exp(1-x) + 7*math.log(x)

Fibonacci_search(f, a0=1, b0=2, tolerance=0.23, epsilon=0.05)
```

```python
def f(x):
    return pow(x,6) - x - 1

Fibonacci_search(f, a0=1, b0=1.5, tolerance=0.01, epsilon=0.05)
```
